```python
def plot_2d_dataset(data,colors):
    pyplot.figure(figsize(8,8))
    pyplot.scatter(x=list(map(lambda x: x[0], data[0])),y=list(map(lambda x: x[1], data[0])),c=data[1], cmap=colors)
```

```python
noise_cirlces = datasets.make_circles(noise=0.15)
plot_2d_dataset(noise_cirlces,colors)
```

```python
simple_classification_problem = datasets.make_classification(n_features=2, n_informative=1, n_redundant=1, n_clusters_per_class=1,random_state=1)
```

```python
plot_2d_dataset(simple_classification_problem,colors)
```

```python
classification_problem = datasets.make_classification(n_features=2, n_informative=2, n_redundant=0, n_classes=4, n_clusters_per_class=1,random_state=1)
```

```python
four_colors = ListedColormap(['red','blue','green','yellow'])
```

```python
plot_2d_dataset(classification_problem,four_colors)
```
